```python
import pandas as pd
```

```python
# Set the column width
pd.options.display.max_colwidth = 200

#  Read in the dataset.
df = pd.read_csv("../../04-Data-Analysis-Pandas/3/Activities/06-Evr_Crowdfunding_Cleaning/Unsolved/Resources/Crowdfunding_data.csv")
df
```
